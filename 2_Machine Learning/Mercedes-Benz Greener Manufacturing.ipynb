## Apply label encoder.

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
train['X0'] = train['X0'].astype('category')
train['X1'] = train['X1'].astype('category')
train['X2'] = train['X2'].astype('category')
train['X3'] = train['X3'].astype('category')
train['X4'] = train['X4'].astype('category')
train['X5'] = train['X5'].astype('category')
train['X6'] = train['X6'].astype('category')
train['X8'] = train['X8'].astype('category')
```

```python
train.info()
```

```python
train_cat = train.select_dtypes(include=['category'])               
```

```python
train_cat.head()
```

```python
train_le = train_cat.apply(le.fit_transform)
```

```python
train_le.head()
```

## Perform dimensionality reduction.

```python
### numerical
train_num = train.select_dtypes(include="int64")
```

```python
###  we use minmax scaler for numerical values
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
```

```python
train_sc = mn.fit_transform(train_le)
train_sc_df = pd.DataFrame(train_sc, index=train_cat.index, columns=train_cat.columns)
train_sc_df.head(10) 
```

```python
## concatonate the x training data back together
x_train = pd.concat([train_sc_df, train_num],axis=1)
x_train
```

```python
from sklearn.decomposition import PCA
```

```python
# perform pca on x_train
```

```python
pca = PCA(0.95)
```

```python
pca.fit(x_train)
```

```python
x_train_pca = pca.transform(x_train)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
len(pca.explained_variance_ratio_)
```

 A total of 76 components are appropriate to capture 95% of variance in the data

## Predict your test_df values using XGBoost.

```python
y_train = train['y']
```

```python
from xgboost import XGBRegressor
```

### First, we do some hyperparameter tuning before we use the xgboost model to predict values

```python
# fit model no training data
model = XGBRegressor(max_depth = 6, colsample_bytree = 0.7)
model.fit(x_train_pca, y_train)
```

```python
# slice of train
x_train1 = x_train.loc[200:300, :]
```

```python
x_train_pca1 = pca.transform(x_train1)
```

```python
y_pred1 = model.predict(x_train_pca1)
```

```python
y1 = train['y'].loc[200:300]
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y1, y_pred1, multioutput='variance_weighted')
```

```python
mean_squared_error(y1, y_pred1)
```

### Through trial an error we find these to be the best parameters, we now tune these parameters into the model and predict y values
